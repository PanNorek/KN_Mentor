```python
import pandas as pd
import numpy as np
```

```python
filename = "data/4.-Badly-Structured-Sales-Data-4.xlsx"
messy_data = pd.read_excel(filename)
```

```python
messy_data.head()
```

```python
messy_data.columns
```

```python
messy_data.columns = ['Order ID', 'Order Date', 'First Class_Consumer', 'First Class_Corporate', 'First Class_Home Office',
       'Same Day_Consumer', 'Same Day_Corporate', 'Same Day_Home Office', 'Second Class_Consumer', 'Second Class_Corporate',
       'Second Class_Home Office', 'Standard Class_Consumer', 'Standard Class_Corporate', 'Standard Class_Home Office']
```

```python
messy_data.drop([0,1], inplace=True)
messy_data.set_index(['Order ID','Order Date'], inplace=True)
```

```python
messy_data.fillna(0, inplace=True)
messy_data.apply(pd.to_numeric)
```

```python
messy_data['Segment'] = messy_data.apply(lambda x: messy_data.columns[x.argmax()], axis = 1)
messy_data["Total"] = messy_data.sum(axis=1)
```

```python
messy_data.columns
```

```python
messy_data.drop(['First Class_Consumer', 'First Class_Corporate',
       'First Class_Home Office', 'Same Day_Consumer', 'Same Day_Corporate',
       'Same Day_Home Office', 'Second Class_Consumer',
       'Second Class_Corporate', 'Second Class_Home Office',
       'Standard Class_Consumer', 'Standard Class_Corporate',
       'Standard Class_Home Office'], axis=1, inplace=True)
```

```python
segments = messy_data["Segment"].str.split("_")
segments = segments.to_list()
messy_data["Segment"] = [segments[i][0] for i in range(len(segments))]
messy_data["Class"] = [segments[i][1] for i in range(len(segments))]
```

```python
messy_data.reset_index(level=0, inplace=True)
```

```python
messy_data["Order Date"] = pd.to_datetime(messy_data["Order Date"].iloc[:-1])
```

```python
messy_data.set_index(["Order Date",'Order ID'], inplace=True)
```
