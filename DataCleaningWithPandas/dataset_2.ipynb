```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = "data/2.-Badly-Structured-Sales-Data-2.xlsx"
messy_data = pd.read_excel(filename)
```

```python
messy_data.head()
```

```python
messy_data.columns
```

```python
messy_data.columns = ['Order Date', 'First Class_Consumer', 'First Class_Corporate', 'First Class_Home Office', 'SameDay_Consumer',
       'SameDay_Corporate', 'SameDay_Home Office', 'SecondClass_Consumer', 'SecondClass_Corporate', 'SecondClass_Home Office',
       'StandardClass_Consumer', 'StandardClass_Corporate', 'StandardClass_Home Office']
```

```python
messy_data.drop([0,1], inplace=True)
```

```python
messy_data.set_index('Order Date', inplace=True)
```

```python
messy_data.fillna(0, inplace=True)
```

```python
messy_data.apply(pd.to_numeric)
```

```python
messy_data['Segment'] = messy_data.apply(lambda x: messy_data.columns[x.argmax()], axis = 1)
```

```python
messy_data["Total"] = messy_data.sum(axis=1)
```

```python
messy_data.drop(['First Class_Consumer', 'First Class_Corporate',
       'First Class_Home Office', 'SameDay_Consumer', 'SameDay_Corporate',
       'SameDay_Home Office', 'SecondClass_Consumer', 'SecondClass_Corporate',
       'SecondClass_Home Office', 'StandardClass_Consumer',
       'StandardClass_Corporate', 'StandardClass_Home Office'],axis=1, inplace=True)
```

```python
segments = messy_data["Segment"].str.split("_")
segments = segments.to_list()
messy_data["Segment"] = [segments[i][0] for i in range(len(segments))]
messy_data["Class"] = [segments[i][1] for i in range(len(segments))]
```

```python
messy_data = messy_data[["Segment","Class","Total"]]
```
