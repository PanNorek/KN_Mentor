```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
messy_data = pd.read_excel("data/1.-Badly-Structured-Sales-Data-1.xlsx")
messy_data.head()
```

```python
messy_data.columns
```

```python
messy_data.columns = ['Order ID', 'Consumer_First Class', 'Consumer_Same Day', 'Consumer_Second Class', 'Consumer_Standard Class',
       'Consumer Total', 'Corporate_First Class', 'Corporate_Same Day',  'Corporate_Second Class','Corporate_Standard Class',
       'Corporate Total', 'HomeOffice_First Class', 'HomeOffice_Same Day', 'HomeOffice_Second Class',
       'HomeOffice_Standard Class', 'Home Office Total']
```

```python
messy_data.fillna(0, inplace=True)
```

```python
messy_data["Segment"] = "0"
```

```python
cols = messy_data.columns.to_list()
messy_data = messy_data[cols[-1:] + cols[:-1]]
messy_data.set_index("Order ID", inplace=True)
```

```python
messy_data.drop(["Ship Mode>>","Order ID"] ,inplace=True)
```

```python
messy_data.head()
```

```python
messy_data = messy_data.drop(["Consumer Total","Corporate Total","Home Office Total"],axis=1)
messy_data.head()
```

```python
messy_data = messy_data.apply(pd.to_numeric)
```

```python
messy_data.info()
```

```python
messy_data['Segment'] = messy_data.apply(lambda x: messy_data.columns[x.argmax()], axis = 1)
```

```python
messy_data.head()
```

```python
messy_data["Total"] = messy_data.sum(axis=1)
```

```python
messy_data.drop(['Consumer_First Class', 'Consumer_Same Day',
       'Consumer_Second Class', 'Consumer_Standard Class',
       'Corporate_First Class', 'Corporate_Same Day', 'Corporate_Second Class',
       'Corporate_Standard Class', 'HomeOffice_First Class',
       'HomeOffice_Same Day', 'HomeOffice_Second Class',
       'HomeOffice_Standard Class'], axis=1, inplace=True)
```

```python
messy_data.shape
```

```python
segments = messy_data["Segment"].str.split("_")
```

```python
segments = segments.to_list()
```

```python
messy_data["Segment"] = [segments[i][0] for i in range(len(segments))]
messy_data["Class"] = [segments[i][1] for i in range(len(segments))]
```

```python
messy_data = messy_data[["Segment","Class","Total"]]
```

```python
messy_data.head()
```

```python
segments = messy_data.groupby(["Segment"])["Total"].sum().reset_index()
segments
```
